```python
diagnosis_counts= df['icd9_code'].value_counts()
top_diagnoses = diagnosis_counts.nlargest(10)
top_diagnoses
```

```python
# Function to retrieve diagnosis description
def get_diagnosis_description(icd9_code, d_icd_diagnoses_df):
    if d_icd_diagnoses_df is not None:
        try:
            return d_icd_diagnoses_df[d_icd_diagnoses_df['icd9_code'] == icd9_code]['short_title'].iloc[0]
        except IndexError:
            # Handle cases where ICD-9 code might not be present in description table
            return "Description Not Found"
    else:
        # Handle scenario where diagnosis description table is not available
        return "Diagnosis Description Unavailable"

for icd9_code, count in top_diagnoses.items():
    description = get_diagnosis_description(icd9_code, d_icd_df)
    print(f"{icd9_code}: {count} admissions ({description})")
```

```python
# Create a dictionary mapping icd9_code to short_title
icd9_to_title = dict(zip(d_icd_df['icd9_code'], d_icd_df['short_title']))

# Calculate the top 5 ICD-9 codes for male patients
top_5_icd_male = df[df['gender'] == 'M']['icd9_code'].map(icd9_to_title).value_counts().nlargest(5)

# Calculate the top 5 ICD-9 codes for female patients
top_5_icd_female = df[df['gender'] == 'F']['icd9_code'].map(icd9_to_title).value_counts().nlargest(5)

# Display the top 5 ICD-9 codes with short titles for each gender
print("Top 5 ICD-9 Codes for Male Patients:")
print(top_5_icd_male)
print("\nTop 5 ICD-9 Codes for Female Patients:")
print(top_5_icd_female)
```

```python
# Calculate the top 5 diagnoses for each admission type
top_diagnoses_by_admission_type = {}
for admission_type in df['admission_type'].unique():
    diagnoses = df[df['admission_type'] == admission_type]['icd9_code'].map(icd9_to_title)
    top_diagnoses_by_admission_type[admission_type] = diagnoses.value_counts().nlargest(5)

# Display the top 5 diagnoses for each admission type
for admission_type, top_diagnoses in top_diagnoses_by_admission_type.items():
    print(f"Top 5 Diagnoses for {admission_type}:")
    print(top_diagnoses)
    print()
```

```python
# Group the data by subject_id and count the unique diagnoses for each patient
patient_diagnoses_count = df.groupby('subject_id')['icd9_code'].nunique()

# Identify patients with multiple diagnoses
patients_with_multiple_diagnoses = patient_diagnoses_count[patient_diagnoses_count > 1].index

# For each patient with multiple diagnoses, list out their exact comorbidities
for patient_id in patients_with_multiple_diagnoses:
    patient_data = df[df['subject_id'] == patient_id]
    patient_diagnoses = patient_data['icd9_code'].map(icd9_to_title).unique()
    print(f"Patient {patient_id} has the following comorbidities:")
    for diagnosis in patient_diagnoses:
        print(f"- {diagnosis}")
    print()
```

```python
def calculate_readmission_rates(df, time_frames):
    readmission_rates = {}
    for time_frame in time_frames:
        # Calculate readmission within the time frame
        readmissions_within_time_frame = df[df['LOS'] <= time_frame]['hadm_id'].nunique()

        # Calculate total discharges
        total_discharges = df['hadm_id'].nunique()

        # Calculate readmission rate
        readmission_rate = readmissions_within_time_frame / total_discharges * 100

        # Store readmission rate
        readmission_rates[f'{time_frame}-Day'] = readmission_rate

    return readmission_rates

# Specify time frames for readmission rates (e.g., 30, 60, 90 days)
time_frames = [30, 60, 90]

# Calculate readmission rates
readmission_rates = calculate_readmission_rates(df, time_frames)

# Print readmission rates within specified time frames
print("Readmission Rates:")
for time_frame, rate in readmission_rates.items():
    print(f"{time_frame}: {rate:.2f}%")
```

```python
# Calculate in-hospital mortality rates based on patient demographics
mortality_rates_by_demographics = df.groupby('gender')['hospital_expire_flag'].mean() * 100

# Calculate in-hospital mortality rates based on types of admissions
mortality_rates_by_admission_type = df.groupby('admission_type')['hospital_expire_flag'].mean() * 100

# Calculate in-hospital mortality rates based on diagnoses
mortality_rates_by_icd9_code = df.groupby('icd9_code')['hospital_expire_flag'].mean() * 100

# Sort the mortality rates in descending order
mortality_rates_by_icd9_code = mortality_rates_by_icd9_code.sort_values(ascending=False)

# Create a DataFrame to include ICD-9 codes, short titles, and mortality rates
mortality_rates_by_icd9_code = pd.DataFrame({
    'icd9_code': mortality_rates_by_icd9_code.index,
    'mortality_rate': mortality_rates_by_icd9_code.values
})

# Assuming 'icd9_to_title' is a dictionary mapping ICD-9 codes to their short titles
# Map short titles to ICD-9 codes
mortality_rates_by_icd9_code['short_title'] = mortality_rates_by_icd9_code['icd9_code'].map(icd9_to_title)



# Print mortality rates based on patient demographics
print("Mortality Rates by Patient Demographics:")
print(mortality_rates_by_demographics)

# Print mortality rates based on types of admissions
print("\nMortality Rates by Admission Type:")
print(mortality_rates_by_admission_type)

# Print mortality rates based on diagnoses
print("\nMortality Rates by Diagnosis:")
print(mortality_rates_by_icd9_code.head(15))
```
